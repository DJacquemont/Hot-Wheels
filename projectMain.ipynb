```python
import cv2
import numpy as np
import math

import vision
```

```python
vid = cv2.VideoCapture(0)
```

```python
while(True):
    terrain = vision.terrainFetch(vid)
    if type(terrain) != bool:
        nodes, nodeCon, maskObsDilated = terrain
        break

while(True):
    output = vision.liveFeedback(vid, nodes, nodeCon, maskObsDilated)
    if type(output) != bool:
        cv2.imshow('frame', output)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        cv2.destroyAllWindows()
        break
```

```python
while(True):
    img = vision.frameReader(vid)
    cv2.imshow('frame', img)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        cv2.destroyAllWindows()
        break
```
